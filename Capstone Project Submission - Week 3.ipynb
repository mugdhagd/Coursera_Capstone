```python
#Extracting table with postal codes for Canada from table
tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
raw_table = tables[0]
print(raw_table)
```

```python
#Cleaning the table and storing in pandas dataframe
df = pd.DataFrame(raw_table)
header = df.iloc[0]
df = df[1:]
df.columns = header
df.drop(df[df['Borough']=='Not assigned'].index, inplace = True)
df.reset_index(inplace=True, drop = True)
df.head()
```

```python
#Concatenating neighbourhoods having identical postcodes
df['Neighbourhood'] = df.groupby(['Postcode'])['Neighbourhood'].transform(lambda x: ', '.join(x))
```

```python
#Replacing 'Not assigned' neighbourhoods with names of corresponding boroughs
df['Neighbourhood'].replace(to_replace = 'Not assigned', value = df['Borough'], inplace = True)
```
